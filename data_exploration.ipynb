```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
recipes = pd.read_excel('Dish_Recipes_full_dataset_v1_0.xlsx')
foods = pd.read_excel('Food_Compass_Score_Food_Items_v1_0.xlsx')
display(recipes.head())
display(foods.head())
```

```python
foods.shape, recipes.shape
```

```python
food_groups_agg = foods.groupby('Group').count().reset_index()
food_groups_agg = food_groups_agg[['Group','foodcode']].rename({'foodcode':'count'},axis=1)
food_groups_agg['pct'] = food_groups_agg['count'] / foods.shape[0]

food_scores_agg = foods.groupby('Food Compass Score').count().reset_index()
food_scores_agg = food_scores_agg[['Food Compass Score','foodcode']].rename({'foodcode':'count'},axis=1)
food_scores_agg['pct'] = food_scores_agg['count'] / foods.shape[0]

fig1 = px.bar(food_scores_agg,x='Food Compass Score',y='count',height=300,title = 'Foods by Compass Score')
fig1.show()

fig2 = px.bar(food_groups_agg,x='Group',y='count',height=300,title = 'Foods by Group')
fig2.show()
```
